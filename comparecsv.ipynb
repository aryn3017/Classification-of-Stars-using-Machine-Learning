```python
import pandas as pd

# Load the two CSV files
csv1 = pd.read_csv('prediction__21324004.csv')
csv2 = pd.read_csv('final_predictions.csv')

# Align columns for comparison
csv1_aligned = csv1.rename(columns={'Predicted_Star_Type': 'Star_Type'})
csv2_aligned = csv2.copy()

# Ensure both have the same shape and columns after alignment
if csv1_aligned.shape != csv2_aligned.shape or not all(csv1_aligned.columns == csv2_aligned.columns):
    raise ValueError("CSV files do not have the same shape or columns even after alignment.")

# Compare each row for exact match
similarity = (csv1_aligned == csv2_aligned).all(axis=1).astype(int)

# Add result to a new DataFrame for inspection
result = csv1.copy()
result['is_identical'] = similarity

# Show the result
print(result[['is_identical']])

# Optionally, save the result
result.to_csv('rowwise_similarity.csv', index=False)
```

```python
percentage_equal = similarity.mean() * 100
print(f"Percentage of identical rows: {percentage_equal:.2f}%")
```
